```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
```

```python
(X_train, y_train), (X_test,y_test) = datasets.cifar10.load_data()
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_train[:5]
```

```python
y_train = y_train.reshape(-1,)
y_train[:5]
```

-1 in first position indicates not to change this first dimension, second place blank indicates to convert the number in array into normal number i.e. flattening it

```python
y_test = y_test.reshape(-1,)
```

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

```python
def plot_sample(X, y, index):
    plt.figure(figsize = (15,2)) #controls the size of image 
    plt.imshow(X[index])
    plt.xlabel(classes[y[index]]) # addind x label
```

defined an image showing function

```python
plot_sample(X_train, y_train, 0)
```

```python
plot_sample(X_train, y_train, 1)
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

Normalizing the training data i.e. converting it in range 0 to 1

whenever we have image, pixel value varies from 0 to 255...hence we divided by 255 to convert in 0 to 1 range

```python
ann = models.Sequential([
        layers.Flatten(input_shape=(32,32,3)),
        layers.Dense(3000, activation='relu'),
        layers.Dense(1000, activation='relu'),
        layers.Dense(10, activation='softmax')    
    ])

ann.compile(optimizer='SGD',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

ann.fit(X_train, y_train, epochs=5)
```

building ANN

```python
from sklearn.metrics import confusion_matrix , classification_report
import numpy as np
y_pred = ann.predict(X_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```

```python
cnn = models.Sequential([
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)), #kernel size is filter size
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

building CNN

in middle layer we don't need to specify input size because the network figures it out itself

softmix used rather than sigmoid because softmix normalise probability i.e. sum of outputs = 1

CNN will figure out filters for us automatically, we just need to specify number of filters, filter size, activation value

```python
cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(X_train, y_train, epochs=10)
```

```python
cnn.evaluate(X_test,y_test)
```

```python
y_pred = cnn.predict(X_test)
y_pred[:5]
```

this outputs probability of each index

```python
y_classes = [np.argmax(element) for element in y_pred]
y_classes[:5]
```

argmax returns index of max probability output
