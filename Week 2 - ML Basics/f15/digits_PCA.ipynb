```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df.describe()
```

```python
X = df
y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

here we used standarscaler , minmax can also be used

scaling before PCA

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

using PCA to reduce dimensions


```python
X
```

```python
# Using components such that 95% of variance is retained
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(X)
X_pca.shape
```

here only 29 columns are left

it calcu;ates new columns, not selecting 29 columns from og dataset

PCA(0.95) this means retain 95% of useful featuresd, hence it can create any number of PCs but it will ensure 95% features data is retained

```python
pca.explained_variance_ratio_
```

it shows that PC1 captures 14% of data/variance, PC2 captures 13% and so on

```python
pca.n_components_
```

PCA created 29 components out of 64 original columns

```python

X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

specify max iter for gradient sescent

```python
# selecting only two components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

We can see that both combined retains 0.14+0.13=0.27 or 27% of important feature information

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
